```python
import pandas as pd
import requests
import re
import unicodedata
import os
```

```python
response = requests.get('https://docs.google.com/spreadsheets/d/1TaFZwPyKXRcLBOmhbpRBFD2f2N3u5cOeuViyQkkxZ3A/export?format=csv')
```

```python
with open('spreadsheet.csv', 'wb') as f:
    f.write(response.content)
```

```python
df = pd.read_csv('spreadsheet.csv')
```

```python
df = df[df['Name'] == df['Name']].copy()
```

```python
def removeAccents(name):
    nfkd_form = unicodedata.normalize('NFKD', name)
    name = u"".join([c for c in nfkd_form if not unicodedata.combining(c)])
    return name
```

```python
def getImageURL(name):
    if name.find('OK') >= 0 or name.find('Ok') >= 0:
        name = re.sub('\(|\)|O[Kk]', '', name)
        return name.strip()
    else:
        return 'missing.jpg'

#df['Image'].apply(getImageURL)
```

```python
def fileExists(name):
    if os.path.isfile(os.path.join('img_cr_resized/JPEG/', name)):
        return 'yes'
    else:
        return 'no'
```

```python
df = df[df['Image'] == df['Image']].copy()
```

```python
df['Image_URL'] = df['Image'].apply(getImageURL)
```
